```python
# Read the file

## Save the path

p = "/content/Base de datos salud corregido (1).xlsx"
## Read and create the dataframe

d = pd.read_excel(p)
d.info()
```

```python
# Explore
d.head()
```

# **3.Cleaning**

## Column names

```python
# Time to time
d.rename(columns= {"Time":"year"}, inplace= True)

# Country Name to country
d.rename(columns= {"Country Name":"country"}, inplace= True)

# Time code to codet
d.rename(columns= {"Time Code":"codet"}, inplace= True)

# Country code to code
d.rename(columns= {"Country Code":"code"}, inplace= True)

# Current health expenditure per capita to current_h
d.rename(columns= {"Current health expenditure per capita (current US$) [SH.XPD.CHEX.PC.CD]":"current_h"}, inplace= True)

#  Death rate, crude (per 1,000 people) [SP.DYN.CDRT.IN] to death
d.rename(columns= {"Death rate, crude (per 1,000 people) [SP.DYN.CDRT.IN]":"death"}, inplace= True)

# Population growth (annual %) [SP.POP.GROW]   to population_g
d.rename(columns= {"Population growth (annual %) [SP.POP.GROW]":"population_g"}, inplace= True)



d.head()
```

## Delete columns

```python
d = d.drop(columns=["codet"])
d.info()
```

# **4. Indexing**

```python
# Duplicate the dataframe
d2 = d
```

```python
# Duplicate year as string / How to create a column
d2["year_str"]  = d2["year"].astype(str)
d2.info()
```

```python
# Indexing
idx = pd.MultiIndex.from_product([d2["country"].unique(),range(d2.year.min(),d2.year.max()+1)],
                                 names = ["country","year"])

## Attach the index
d2 = d2.set_index(["country","year"])


d2.head()
```

# **5. Delete Nan ..**

```python
d2.isna()
```

```python
# By the flies
d2 = d2.replace("..",np.NaN)
d2.head()
```

```python
# Delete Nan
d2 = d2.dropna()
d2.info()
```

```python
# By the flies
d2 = d2.replace("nan",np.NaN)
d2.head()
```

```python
# Delete Nan
d2 = d2.dropna()
d2.info()
```

# **9. Another graphs**

(One) line graph

```python
france =d2.filter(like = "France", axis=0)
france.head()
```

```python
# Center the title
fig1= px.line(france, x = "year_str", y = "death")
fig1.update_layout(title_text = 'Death in France through the years',
                  title_font_size = 30,
                  title_x = 0.5)

# Modify the text in the axis
fig1.update_xaxes(title = 'Years', title_font_size = 25,
                 dtick = "M1",
                 tickformat = "%b\n%Y")
fig1.update_yaxes(title = 'Death', title_font_size = 25)

fig1.show()

```

HISTORIA

```python
## Barras
# Center the title
fig2= px.bar(france, x = "year_str", y = "death")
fig2.update_layout(title_text = 'Death in France through the years',
                  title_font_size = 30,
                  title_x = 0.5)

# Modify the text in the axis
fig2.update_xaxes(title = 'Years', title_font_size = 25,
                 dtick = "M1",
                 tickformat = "%b\n%Y")
fig2.update_yaxes(title = 'Death', title_font_size = 25)

fig2.show()
```

```python
## Libraries
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
## Crear a figure with secondary axis

fig3= make_subplots(specs = [[{"secondary_y": True}]])

## Add each graph
fig3.add_trace(
    go.Scatter(x = france.year_str, y = france.death, name = " Death"),
    secondary_y = False,
)

## Add each graph
fig3.add_trace(
    go.Scatter(x = france.year_str, y = france.current_h, name = " current health"),
    secondary_y = True,
)

```
