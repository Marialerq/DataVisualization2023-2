# **3.Cleaning**

## Column names

```python
# Time to time
d.rename(columns= {"Time":"year"}, inplace= True)

# Country Name to country
d.rename(columns= {"Country Name":"country"}, inplace= True)

# Time code to codet
d.rename(columns= {"Time Code":"codet"}, inplace= True)

# Country code to code
d.rename(columns= {"Country Code":"code"}, inplace= True)

# Current health expenditure per capita to current_h
d.rename(columns= {"Current health expenditure per capita (current US$) [SH.XPD.CHEX.PC.CD]":"current_h"}, inplace= True)

#  Death rate, crude (per 1,000 people) [SP.DYN.CDRT.IN] to death
d.rename(columns= {"Death rate, crude (per 1,000 people) [SP.DYN.CDRT.IN]":"death"}, inplace= True)

# Population growth (annual %) [SP.POP.GROW]   to population_g
d.rename(columns= {"Population growth (annual %) [SP.POP.GROW]":"population_g"}, inplace= True)



d.head()
```

## Delete columns

```python
d = d.drop(columns=["codet"])
d.info()
```
